```python
cross= pd.crosstab(database.estado, database.rede_ensino, margins=True,rownames=['Estado'], colnames=['Rede_de_Ensino'])

cross.index = state_index
cross.columns = [
    'Pública',
    'Privada',
    'Total']

cross
```

```python
cross = pd.crosstab(database.estado, database.nível_de_escolaridade, margins=True, rownames=['Estado'], colnames=['Nível_de_Escolaridade'])

cross.index = state_index

cross.columns = ['Maternal, jardim de infância etc.',
'Classe de alfabetização',
'Alfabetização de adultos',
'1ª a 4ª séries do ensino fundamental regular em 8 anos ou de educação de jovens e adultos',
'5ª a 8ª séries do ensino fundamental regular em 8 anos ou de educação de jovens e adultos',
'1ª série do ensino fundamental regular em 9 anos',
'2ª a 5ª séries do ensino fundamental regular em 9 anos',
'6ª a 9ª séries do ensino fundamental regular em 9 anos',
'Educação de jovens e adultos não seriado do ensino fundamental',
'Ensino médio, inclusive pré-vestibular',
'Superior, inclusive mestrado ou doutorado',
'Total'
]

cross
```

```python
def percConvert(tab):
  key= 6293+1073+435+1092+2162+5685+25397+22972+473+17403+13644+96629
  return (tab*100/float(tab[-1])).round(decimals=0)

cross= cross.T.apply(percConvert, axis=0)
cross= cross.T

plot1= cross.plot(kind='bar', stacked=True, legend=False)
plot1.set_ylim([0, 100])
plt.ylabel('Freq.Relativa (em%)')

plot2= cross.plot(kind='bar', stacked=True)
plot2.set_ylim([0, 100])
```
